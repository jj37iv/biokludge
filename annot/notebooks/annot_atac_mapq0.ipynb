```python
# atac wt_pe, wt_se, glp1_se: merge across tissues
df_wt_pe = read_gffbed('annot_ce10_mapq0/metrics_atac/atac_wt_pe_0.001.bed')
df_wt_se = read_gffbed('annot_ce10_mapq0/metrics_atac/atac_wt_se_0.001.bed')
df_glp1_se = read_gffbed('annot_ce10_mapq0/metrics_atac/atac_glp1_se_0.001.bed')

df_wt_pe['atac_source'] = 'atac_wt_pe_mapq0'
df_wt_se['atac_source'] = 'atac_wt_se_mapq0'
df_glp1_se['atac_source'] = 'atac_glp1_se_mapq0'


print('%d peaks in wt_pe' % (len(df_wt_pe),))
print('%d peaks in wt_se' % (len(df_wt_se),))
print('%d peaks in glp1_se' % (len(df_glp1_se),))

bt_wt_se_only = BedTool.from_dataframe(df_wt_se).subtract(b=BedTool.from_dataframe(df_wt_pe), A=True)
df_wt_se_only = pd.read_csv(bt_wt_se_only.fn, sep='\t', names=df_wt_se.columns)
df_wt_either = pd.concat([df_wt_pe, df_wt_se_only], axis=0, ignore_index=True).sort_values(yp.NAMES_BED3).reset_index(drop=True)
print('%d peaks in wt_mapq0_either' % (len(df_wt_either),))

bt_glp1_se_only = BedTool.from_dataframe(df_glp1_se).subtract(b=BedTool.from_dataframe(df_wt_either), A=True)
df_glp1_se_only = pd.read_csv(bt_glp1_se_only.fn, sep='\t', names=df_wt_se.columns)
print('%d peaks in glp1_mapq0_se' % (len(df_glp1_se_only),))

df_wt_glp1 = pd.concat([df_wt_either, df_glp1_se_only], axis=0, ignore_index=True).sort_values(yp.NAMES_BED3).reset_index(drop=True)
print('%d peaks in df_wt_glp1_mapq0' % (len(df_wt_glp1),))
```
