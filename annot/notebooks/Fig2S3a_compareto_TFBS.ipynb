```python
fp_TFBS = 'annot/Fig2S3_overlaps/modERN_modENCODE/modERN_modENCODE_ext200_merge.bed'
fp_Daugherty2017 = 'annot/Fig2S3_overlaps/Daugherty2017/Daugherty2017_consensus_ATAC.bed'
fp_Ho2017 = 'annot/Fig2S3_overlaps/Ho2017/Ho2017_DNase_emb.bed'

df_TFBS = pd.read_csv(fp_TFBS, sep='\t', names=yp.NAMES_BED9[:4])
df_Daugherty2017 = pd.read_csv(fp_Daugherty2017, sep='\t', comment='#', names=yp.NAMES_BED9)[yp.NAMES_BED3]
df_Ho2017 = pd.read_csv(fp_Ho2017, sep='\t', names=yp.NAMES_BED3)
```
